```python
df = pd.read_csv('ipl.csv')
```

```python
df1 = df.drop(['mid','batsman','bowler','striker','non-striker'],axis='columns')
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['venue'].unique()
```

```python
df1['venue'].nunique()
```

```python
df1['bat_team'].unique()
```

```python
regular_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals', 'Mumbai Indians','Kings XI Punjab','Royal Challengers Bangalore', 'Delhi Daredevils','Sunrisers Hyderabad']
```

```python
df1 = df1[(df1['bat_team'].isin(regular_teams)) & (df1['bowl_team'].isin(regular_teams))]
```

```python
df1['bat_team'].unique()
```

```python
df1.head()
```

```python
df1['venue'].nunique()
```

```python
from datetime import datetime
```

```python
df1['date'] = df1['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
```

```python
df1.head()
```

```python
df2 = pd.get_dummies(data=df1, columns=['bat_team','bowl_team'])
```

```python
df2.head()
```

```python
df3 = pd.get_dummies(df2['venue'])
```

```python
data = pd.concat([df2,df3],axis='columns')
```

```python
pd.set_option('display.max_columns', None)
```

```python
data.head()
```

```python
data = data.drop('venue',axis='columns')
```

```python
data.head()
```

```python
data.columns
```

```python
data = data[data['overs']>=6.0]
```

```python
data.head()
```

```python
X_train = data.drop(labels='total', axis='columns')[data['date'].dt.year <= 2016]
X_test = data.drop(labels='total', axis='columns')[data['date'].dt.year >= 2017]
```

```python
y_train = data[data['date'].dt.year <= 2016]['total'].values
y_test = data[data['date'].dt.year >= 2017]['total'].values
```

```python
X_train.drop(labels='date', axis='columns', inplace=True)
X_test.drop(labels='date', axis='columns', inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
import pickle
```

```python
filename = 'ipl-prediction.pkl'
pickle.dump(lr, open(filename, 'wb'))
```
